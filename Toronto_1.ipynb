```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
toronto_data = pd.read_html(url)
toronto_data

df = toronto_data[0]

df2 = df[['Postal Code', 'Borough', 'Neighbourhood']]
print(df2)
```

The dataframe will consist of three columns: Postal Code, Borough, and Neighbourhood.

```python
toronto_neighbourhoods = pd.DataFrame(df2)
toronto_neighbourhoods
```

Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
indexNames = toronto_neighbourhoods[toronto_neighbourhoods['Borough'] == 'Not assigned'].index
toronto_neighbourhoods.drop(indexNames, inplace = True)
toronto_neighbourhoods
```

If a cell has a borough but a Not assigned neighbourhood, the neighbourhood should be the same as the borough.

Neighbourhoods that share a postal code should appear in one row, separated with a comma.

```python
toronto_neighbourhoods.loc[toronto_neighbourhoods['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = toronto_neighbourhoods['Borough']
toronto_neighbourhoods
```
